### reading in transmission data

```python
data = scipy.io.loadmat('../Wen/'+file_mat+'.mat')
t = data['TM']
lambdas = data['wavelength'][0,:]

t = t[:,:1,:]
```

```python
print "# pixels: %i \n# input angles: %i \n# wavelengths: %i" % np.shape(t)  

Npixel = np.shape(t)[0]
Npixel_xy = np.sqrt(Npixel)

Nlambda = np.shape(lambdas)[0]
lambda_range = lambdas[-1] - lambdas[0]
dlambda = lambdas[1] - lambdas[0]

lambdas_shift = np.arange(Nlambda)*dlambda

wlambdas = np.linspace(0.0, 2.0*np.pi/dlambda, Nlambda)
wlambda_range = wlambdas[-1] - wlambdas[0]
dwlambda = 2.0 * np.pi / lambda_range
```

### perform Fourier transformation

```python
t_ft = np.fft.fftshift(np.fft.fft(t, axis=2), axes=2)

t_ft_abs = np.abs(t_ft)

t_pic = np.reshape(t, (Npixel_xy, Npixel_xy, 1, Nlambda))
```

### rolling data to filter linear slope in phase

```python
spec = np.mean(t_ft_abs[:,0,:]**2, axis=0)

Norm_spec = np.sum(spec)
spec /= Norm_spec

indx_roll = 0
plot(wlambdas, spec, '-b')

for i in range(5):
    wl_mean = np.einsum('w,w', wlambdas, spec)
    indx_roll_inc = int((np.mean(wlambdas) - wl_mean) / dwlambda)
    indx_roll += indx_roll_inc
    spec = np.roll(spec, indx_roll_inc, axis=0)

t_ft_roll = np.roll(t_ft, indx_roll, axis=2)

plot(wlambdas, spec, '-g')
print wl_mean, np.mean(wlambdas)
```

### perform inverse Fourier transform, separate into amplitude and phase

```python
t_roll = np.fft.ifftshift(np.fft.ifft(t_ft_roll, axis=2), axes=2)

t_roll_amp = np.abs(t_roll)
t_roll_phase = np.unwrap(np.angle(t_roll))

print np.sum(np.abs(t_roll) - np.abs(t)) # should be zero
```

```python
np.save("t_roll", t_roll)
```

### define which pixels to transform (chosen and next one)

```python
#pixel = 204 + Npixel_xy * 182
pixel = 65 + Npixel_xy * 60


t_pixel = t_roll[pixel:pixel+1000]
t_pixel_amp = t_roll_amp[pixel:pixel+1000]
t_pixel_phase = t_roll_phase[pixel:pixel+1000]

k_choice = 0
pixel_choice = 0
```

### perform Fourier transform for amplitude and phase

```python
t_pixel_ft = np.fft.fftshift(np.fft.fft(t_pixel, axis=2), axes=2)
t_pixel_ft = np.fft.fft(t_pixel, axis=2)

t_pixel_amp_ft = np.fft.fftshift(np.fft.fft(t_pixel_amp, axis=2), axes=2)
t_pixel_phase_ft = np.fft.fftshift(np.fft.fft(t_pixel_phase, axis=2), axes=2)
```

### intensity correlation function

```python
# add average over different lambdas, atm just for lambda_min
intensity = np.abs(t)**2

intensity_0 = intensity[:,:,0]

mean_int = np.mean(intensity, axis=2)

reldev0 = intensity_0 / mean_int - 1.0
reldevl = np.einsum('xkl,xk->xkl', intensity, 1.0/mean_int) - 1.0

iCorr = np.mean(np.einsum('xk,xkl->xkl', reldev0, reldevl), axis=0)
iCorr = np.einsum('kl,k->kl', iCorr, 1.0 / iCorr[:,0])
drop_mask = iCorr <= 0
iCorr_drop = cp.deepcopy(iCorr)
iCorr_drop[drop_mask] = 0.0
iCorr_drop = np.einsum('kl,k->kl', iCorr_drop, 1.0 / np.sum(iCorr_drop, axis=1))


iCorr_Width_sqr = np.einsum('l,kl->k', lambdas_shift**2, iCorr_drop)
iCorr_Width = np.sqrt(iCorr_Width_sqr)

mean_iCorr_Width = np.mean(iCorr_Width)

print "mean intensity correlation width: ", mean_iCorr_Width

#plot(np.arange(Nlambda)*dlambda, iCorr[k_choice])
plot(np.arange(Nlambda)*dlambda, iCorr_drop[k_choice])
```

### filter data

```python
mean_power = np.mean(np.abs(t_pixel_ft[:,k_choice,:])**2, axis=0)
N_ft = np.sum(mean_power)

av_ft = np.einsum('w,w', wlambdas, mean_power / N_ft)
av_sqr_ft = np.einsum('w,w', wlambdas**2, mean_power / N_ft)
spread_ft = np.sqrt(av_sqr_ft - av_ft**2)

#filter_params = (4.0*spread_ft, wlambda_range/300.0)
filter_params = (2.0 * 2.0 * np.pi / (2.0 * mean_iCorr_Width),
                 wlambda_range/300.0)


N_fermi = np.sum(dpu.fermi(wlambdas - av_ft, filter_params))
plot(wlambdas,
     dpu.fermi(wlambdas - av_ft, filter_params) * N_ft/N_fermi,
     '-r', lw=2.0)

import copy

#ylim(0,1e9)
indx_dc = np.argmax(mean_power)
mean_power_no_dc = copy.deepcopy(mean_power)
mean_power_no_dc[indx_dc-2:indx_dc+3] = 0.0

#plot(wlambdas, mean_power[:], 'd-b')
plot(wlambdas, mean_power_no_dc[:], 'd-g')
#plot(np.roll(wlambdas, len(wlambdas)/2), mean_power_no_dc[:], 'd-b')

```

```python
periods = 2.0 * np.pi / (wlambdas - np.mean(wlambdas))

mean_power_amp = np.mean(np.abs(t_pixel_amp_ft[:,k_choice,:])**2, axis=0)
N_amp_ft = np.sum(mean_power_amp)

av_amp_ft = np.einsum('w,w', wlambdas, mean_power_amp / N_amp_ft)
av_sqr_amp_ft = np.einsum('w,w', wlambdas**2, mean_power_amp / N_amp_ft)
spread_amp_ft = np.sqrt(av_sqr_amp_ft - av_amp_ft**2)

#filter_params_amp = (5.0*spread_amp_ft, wlambda_range/300.0)
filter_params_amp = filter_params


N_fermi = np.sum(dpu.fermi(wlambdas - av_amp_ft, filter_params_amp))
plot(wlambdas,
     dpu.fermi(wlambdas - av_amp_ft, filter_params_amp) * N_amp_ft/N_fermi,
     '-r', lw=2.0)

#ylim(0,1e9)
indx_dc = np.argmax(mean_power)
mean_power_amp_no_dc = cp.deepcopy(mean_power_amp)
mean_power_amp_no_dc[indx_dc-2:indx_dc+3] = 0.0

ylim(0,1e5)
#xlim(0,np.max(periods))
#plot(periods, mean_power_amp_no_dc[:], 'd-b')
plot(wlambdas, mean_power_amp[:], 'd-b')
show()

ylim(-1,1)
plot(wlambdas, periods)
```

```python
mean_power_phase = np.mean(np.abs(t_pixel_phase_ft[:,k_choice,:])**2, axis=0)
N_phase_ft = np.sum(mean_power_phase)

av_phase_ft = np.einsum('w,w', wlambdas, mean_power_phase / N_phase_ft)
av_sqr_phase_ft = np.einsum('w,w', wlambdas**2, mean_power_phase / N_phase_ft)
spread_phase_ft = np.sqrt(av_sqr_phase_ft - av_phase_ft**2)

#filter_params_phase = (2.5*spread_phase_ft, wlambda_range/300.0)
filter_params_phase = filter_params

N_fermi = np.sum(dpu.fermi(wlambdas - av_phase_ft, filter_params_phase))
plot(wlambdas,
     dpu.fermi(wlambdas - av_phase_ft, filter_params_phase) * N_phase_ft/N_fermi,
     '-r', lw=2.0)

ylim(0,1e5)
plot(wlambdas, mean_power_phase[:], 'd-b')
```

```python
t_pixel_ft_filter = dpu.low_pass_filter(t_pixel_ft,
                                      wlambdas - av_ft,
                                      dpu.fermi,
                                      filter_params,
                                      False)
t_pixel_amp_ft_filter = dpu.low_pass_filter(t_pixel_amp_ft,
                                      wlambdas - av_amp_ft,
                                      dpu.fermi,
                                      filter_params_amp,
                                      False)
t_pixel_phase_ft_filter = dpu.low_pass_filter(t_pixel_phase_ft,
                                      wlambdas - av_phase_ft,
                                      dpu.fermi,
                                      filter_params_phase,
                                      False)
```

### perform inverse Fourier transform of filtered data

```python
t_pixel_filter = dpu.norm_ifft(t_pixel, t_pixel_ft_filter)
t_pixel_amp_filter = dpu.norm_ifft_shift(t_pixel_amp, t_pixel_amp_ft_filter)
t_pixel_phase_filter = dpu.norm_ifft_shift(t_pixel_phase, t_pixel_phase_ft_filter)
```

### compare original and filtered data

```python
plot(lambdas, np.abs(t_pixel[pixel_choice,k_choice,:]), '-b', lw=1.0)
plot(lambdas, np.abs(t_pixel_filter[pixel_choice,k_choice,:]), '-r', lw=1.0)
show()
clf()

plot(lambdas, np.unwrap(np.angle(t_pixel[pixel_choice,k_choice,:])), '-b', lw=1.0)
plot(lambdas, np.unwrap(np.angle(t_pixel_filter[pixel_choice,k_choice,:])), '-r', lw=1.0)
show()
clf()
```

```python
plot(lambdas, t_pixel_phase[pixel_choice,k_choice,:], 'd-b', lw=1.0)
plot(lambdas, t_pixel_phase_filter[pixel_choice,k_choice,:], '-r', lw=2.0)
```
